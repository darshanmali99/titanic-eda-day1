```python
students = [
    {"name":"Prem", "age":19, "marks":65},
    {"name":"Shubham", "age":21, "marks":70},
    {"name":"Mayur", "age":25, "marks":75},
    {"name":"Krish", "age":23, "marks":90}
]    
```

```python
def check_adult(age):
    if age >=18:
        return"Adult"
    else:
        return"Minor"

def check_pass(marks):
    if marks >=50:
        return"Pass"
    else:
        return"Fail"
```

```python
for student in students:
    name = student["name"]
    age = student["age"]
    marks = student["marks"]

    print(
        name,
        "|",
        check_adult(age),
        "|",
        check_pass(marks),
        "|"
    )
```

```python
result = []
```

```python
result = []

for student in students:
    record = {
        "Name": student["name"],
        "Age group": check_adult(student["age"]),
        "Result": check_pass(student["marks"])
    }
    result.append(record)
```

```python
result
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame(result)
df
```
